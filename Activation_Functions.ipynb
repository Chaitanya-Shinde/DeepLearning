#Exploring different activation functions

1. Sigmoid
2. ReLU
3. Hyperbolic tangent

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

### Sigmoid activation function

```python
#Defining the sigmoid activation function
def sigmoid(z): # here z in the weighted sum of the inputs to the neuron along with the weights and bias
  return 1 / (1 + np.exp(-z))
```

```python
#Defining the derivative of the sigmoid function, i.e. sigmoid gradient
def sigmoidDerivative(z):
  return sigmoid(z) * (1 - sigmoid(z))
```

### ReLU activation function

```python
#Defining the ReLU activation function
def relu(z):
  return np.maximum(0,z)
```

```python
#Defining the derivative of the ReLU activation function, i.e. ReLU gradient
def reluDerivative(z):
  return np.where(z > 0, 1, 0)
```

### Hyperbolic tangent activation function

```python
#Defining the hyperbolic tangent activation function
def tanh(z):
  return np.tanh(z)
```

```python
#Defining the derivatinve of the hyperbolic tangent function, i.e. Hyperbolic tangent gradient
def tanhDerivative(z):
  return 1 - np.tanh(z) ** 2
```

### generating inputs

Generating 400 synthetic values between -10 and 10. The gradient would be calculated for each of these values.


```python
z = np.linspace(-10, 10, 400)
```

Next, let's compute the derivatives for Sigmoid, Relu, and Hyperbolic tangent activations and assign the values to sigmoidGradient, reluGradient and tanhGradient respectively

```python
sigmoidGradient = sigmoidDerivative(z)
reluGradient = reluDerivative(z)
tanhGradient = tanhDerivative(z)
```

```python
plt.figure(figsize=(12,12))

plt.subplot(2,2,1)
plt.plot(z,sigmoid(z), label = "sigmoid activation", color  = 'blue')
plt.plot(z,sigmoidGradient, label = "sigmoid derivative", color = 'red', linestyle = '--')
plt.title('Sigmoid activation and gradient')
plt.xlabel('input value (z)')
plt.ylabel('activation / gradient')
plt.legend()

plt.subplot(2,2,2)
plt.plot(z,relu(z), label = 'Relu activation', color = 'blue')
plt.plot(z,reluGradient, label = 'Relu derivative', color='red', linestyle = '--'  )
plt.title('Relu activation and gradient')
plt.xlabel('input value (z)')
plt.ylabel('activation / gradient')
plt.legend()

plt.subplot(2,2,3)
plt.plot(z,tanh(z), label = 'Hyperbolic tangent activation', color = 'blue')
plt.plot(z,tanhGradient, label = 'Hyperbolic tangent derivative', color='red', linestyle = '--'  )
plt.title('Hyperbolic tangent activation and gradient')
plt.xlabel('input value (z)')
plt.ylabel('activation / gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
